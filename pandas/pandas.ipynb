```python
from typing import TypedDict

class User(TypedDict):
    name:list[str]
    age: list[int]
    city: list[str]
    


user: User = {
    'name': ['Alice','Ama','Kyereh'],
    'age': [12,45,56],
    'city': ['London','Accra', 'NY']
}

df = pd.DataFrame(user)

df
```
